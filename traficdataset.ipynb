```python
import pandas as pd

df['date'] = pd.to_datetime(df['date'])
```

```python
df['year_month'] = df['date'].dt.to_period('M')
```

```python
monthly_trend = (
    df.groupby(['year_month', 'track'])['event']
      .count()
      .reset_index(name='total_events')
)
```

```python
top_tracks_per_month = (
    monthly_trend
    .sort_values(['year_month', 'total_events'], ascending=[True, False])
    .groupby('year_month')
    .head(5)
)
```

```python
for month in top_tracks_per_month['year_month'].unique():
    print(f"\nTop Songs for {month}")
    print(
        top_tracks_per_month[
            top_tracks_per_month['year_month'] == month
        ][['track', 'total_events']]
    )
```

```python
import pandas as pd

# Make sure date column is proper
df['date'] = pd.to_datetime(df['date'])
df['year_month'] = df['date'].dt.to_period('M')

# Create monthly trend table
monthly_trend = (
    df.groupby(['year_month', 'track'])['event']
      .count()
      .reset_index(name='total_events')
)

# 🔹 Take user input
user_month = input("Enter month (YYYY-MM): ")

# Convert to Period
user_month = pd.Period(user_month, freq='M')

# Filter data
result = monthly_trend[monthly_trend['year_month'] == user_month]

# Show top 5
print(result.sort_values('total_events', ascending=False).head(5))
```
